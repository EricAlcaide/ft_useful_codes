## Definició del problema

El primer inidcador  bursàtil que s'ensenya es la mitjana mòbil simple, que és la mitjana dels darrers **x** períodes. També un dels primers indicadors que s'ensenya és el **creuament de mitges**, que es basa en l'assumpció que si agafem una mitja curta que tregui soroll al preu, i una mitja llarga que vagi una mica retardada i, per tant, sigui l'indicador de la "tendència", podem agafar els punts on aquestes dues mitjes es creuin com a indicador d'un canvi de tendència.
El problema que sorgeix és que la mitja llarga va retardada perquè comprèn un gran nombre de períodes anteriors i, per tant, arriba tard als canvis de tendència.

Per tal de solucionar això, ens proposem crear una xarxa convolucional que predigui amb anterioritat els creuamentys de mitjes i, per tant, elimini el retràs que el caracteritza. Es tracta, doncs, d'un **problema de classificació** on haurem de distingir entre si s'apropa un creuament o no. 

```python
# Import libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from keras.models import Sequential, load_model
from keras.layers import Dense, Embedding, Dropout, Flatten
from keras.layers.convolutional import Conv1D
from keras.layers.pooling import MaxPooling1D
```

### Extracció de dades

Extraiem les dades del nostre fitxer .csv i calculem la mitjana curta i la mitjana llarga. El nombre **x** de períodes sobre els quals fan la mitjana ha estat triat de manera totalment arbitrària.

```python
# Data
df = pd.read_csv('GBPJPY_50.csv', delimiter=",").iloc[:, 1].values
# Short Moving Average
window = 100
med = np.array([ np.mean(df[i-window:i+1]) for i in range(window, len(df))])
# Long Moving Average
window2 = 750
med2 = np.array([ np.mean(df[i-window2:i+1]) for i in range(window2, len(df))])
```

```python
print(df.shape, med.shape, med2.shape)
```

Les mitjes no han de perquè coincidir en el mateix punt. Per tant, declararem que és un punt de tall si es troben a menys d'una certa distància escollida de manera arbitrària.

```python
# If distance between two medians is less than treshold: cross point
treshold = 0.005
points = np.array([ 0 if abs(med[window2-window+i]-med2[i]) > treshold else 1 for i in range(len(df)-window2)])
x_coord = np.array([i for i,point in enumerate(points[:10000]) if point == 1])
y_coord = np.array([med2[i] for i in x_coord])
x_coord = np.array([i+window2 for i in x_coord])
```

### Visualització inicial de dades

* En blau podem veure el preu, que té molt de soroll.
* En vermell podem veure la mitja curta.
* En groc podem veure la mitja llarga, que és un indicador més clar de la tendència.
* En verd podem veure els punts que hem marcat com a creuament.

```python
# Initial plot to visualize data
max_plot = min(10000, len(df))

plt.figure()
plt.plot(df[:max_plot])
plt.plot(np.arange(window, max_plot), med[:max_plot-window], 'r')
plt.plot(np.arange(window2, max_plot), med2[:max_plot-window2], 'y')
plt.plot(x_coord, y_coord, '-og')
plt.grid()
plt.show()
```

### Preparació de les dades

Abans d'introduïr les dades al nostre model, hem d'etiquetarles. Com el nostre objectiu és que el model aprengui a predir amb anterioritat el creuament de dues mitjes, etiquetarem un punt com a tall si en els futurs **y** períodes hi ha un tall. El nombre **y** serà tirat de manera arbityrària tenint en compte la quantitat de dades que introduïrem al model.

A més estructurarem les nostres dades en vectors que continguin, per a cada episodi, els 446 últims valors de: el preu, la mitja curta i la mitja llarga. Donats aquests 446·3 valors, demanarem al model que classifiqui el període en 0/1 depenent si s'aproxima un tall de mitjes.

```python
# Create labels
predict_window = 75
y_labeled = np.array([ 1 if points[i:i+predict_window].any() == 1 else 0 for i in range(len(points)-predict_window)])
```

```python
x_data = pd.DataFrame({'prices': df[window2:len(df)-predict_window],
                       'med_s': med[window2-window:len(med)-predict_window]-df[window2:len(df)-predict_window],
                       'med_h': med2[:len(med2)-predict_window]-med[window2-window:len(med)-predict_window]}).as_matrix()[:50000]
win_input = 446
```

```python
x_data = np.array([x_data[i-win_input:i] for i in range(win_input, len(x_data))])
y_labels = y_labeled[win_input:50000]
print(x_data.shape, y_labels.shape)
```

### Creació del model

Per la naturalesa de les nostres dades (seqüencials), les xarxes neuronals "clàssiques" amb capes totalment conectades no ens serveixen, ja que no estem analitzant característiques, sinó tendències o variacions. Per aquesta raó, el nostre model constarà de dos parts:
* Primera part **convolucional**: aquesta part serà l'encarregada d'identificar patrons tant en el preu, com en la mitja curta i la llarga. Aquesta part es compon tant de **blocs convolucionals** on tenen lloc complicades operacions de matrius (convolucions) com per blocks de **Max Pooling**, una tècnica que consisteix a agafar el màxim d'entre **x** valors per tal de reduir la quantitat de dades sense perdre gaire qualitat. Al final d'aquesta part trobem una capa anomenada **Flatten** que transformarà el tensor resultant de la primera part en un únic vector.
* Segona part **clàssica**: aquesta part s'encarregarà de classificar el període segons les característiques identificades per la part convolucional.

```python
# Model Params. 4*4 patches for convolutions. 3 different vectors.
kernel_size,filters = 4,3
# Will take the max from each 2 values.
pooling = 2
shape_0, shape_1 = x_data.shape[1],x_data.shape[2]
```

```python
# Create the model
model = Sequential((
    # The first conv layer learns `nb_filter` filters (aka kernels), each of size ``(filter_length, nb_input_series)``.
    # Its output will have shape (None, window_size - filter_length + 1, nb_filter), i.e., for each position in
    # the input timeseries, the activation of each filter at that position.
    Conv1D(filters=filters, kernel_size=kernel_size, padding='valid', activation='linear', input_shape=(shape_0, shape_1)),
    MaxPooling1D(pool_size = 3),     # Downsample the output of convolution by 2X.
    Conv1D(filters=filters, kernel_size=kernel_size, padding='valid', activation='linear'),
    MaxPooling1D(pool_size = 3),
    Conv1D(filters=filters, kernel_size=kernel_size, padding='valid', activation='linear'),
    MaxPooling1D(pool_size = 3),
    Conv1D(filters=filters, kernel_size=kernel_size, padding='valid', activation='linear'),
    MaxPooling1D(pool_size = 3),
    Conv1D(filters=filters, kernel_size=kernel_size, padding='valid', activation='linear'),
    Flatten(),
    Dense(256, activation="relu"),
    Dense(512, activation="relu"),
    Dense(512, activation="relu"),
    Dense(256, activation="relu"),
    Dense(128, activation="relu"),
    Dense(1, activation='sigmoid'),     # For binary multi-class classification, change to 'softmax'
))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])
```

### Entrenament i Avaluació

Entrenarem el model amb només ~45.000 períodes i l'avaluarem amb 5.000 en els quals no hagi estat entrenat per tal de veure si hi ha algun aprenentatge. Al ser tan profund, pot ocórrer que trigui una mica a "aprendre" i identificar els diferents patrons que després usarà per a la classificació. Un fet que pot accelerar aquesta identificació és entrenar la xarxa amb dades diferents que agafarem del DataFrame que hem creat abans amb el nom de `x_data`.

L'indicador de referència per a l'entrenament serà la precisió (% d'encerts sobre el total) i, en cas que aquesta no millori, la loss, que ha de ser el més petita possible.

```python
# model = load_model('2_batches.h5')
his = model.fit(x_data, y_labels, epochs=5, batch_size=64, verbose=1, validation_split=0.1, shuffle=True)
```

```python
x_data = pd.DataFrame({'prices': df[window2:len(df)-predict_window],
                       'med_s': med[window2-window:len(med)-predict_window]-df[window2:len(df)-predict_window],
                       'med_h': med2[:len(med2)-predict_window]-med[window2-window:len(med)-predict_window]}).as_matrix()[55000:60000]
x_test = np.array([x_data[i-win_input:i] for i in range(win_input, len(x_data))])
# Make predictions on unseen data
predicts = model.predict(x_test)[:-(window2)]
```

```python
%matplotlib notebook
# Plot results
batch, max_plot = 55000,5000
plt.figure()
plt.plot(df[batch:batch+max_plot])
plt.plot(np.arange(window, max_plot), med[batch:batch+max_plot-window], 'r')
plt.plot(np.arange(window2, max_plot), med2[batch:batch+max_plot-window2], 'y')

for prob in [.4, .5, .7, .8, .9]:
    # Catch points above certain prob
    x_coord = np.array([i for i,pred in enumerate(predicts) if pred >= prob])
    y_coord = np.array([df[batch+window2+win_input+x]+0.2*prob for x in x_coord])
    x_coord = np.array([x+window2+win_input for x in x_coord ])
    print(prob, ":", len(x_coord))
    plt.plot(x_coord, y_coord, '-o', alpha =0.4)

plt.plot(np.arange(window2+win_input, max_plot),( predicts*0.75+np.amin(df[batch:batch+max_plot])), alpha=.5)
# Plot graph
plt.legend(['Price', 'Short Median', 'Long Median',
           'prob>0.4', 'prob>0.5', 'prob>0.7', 'prob>0.8',
           'prob>0.9', 'prob*0.75'], loc='upper left')                                                    

plt.grid()  
plt.show()
```
